## Included example code

```python
# example code for Sierpinski Triangle, was included in the Jupyter notebook

#import ColabTurtle.Turtle as t
t.initializeTurtle()
t.bgcolor('white')
t.pencolor('black')
t.speed(13) # 1-13


t.penup()
#t.back(40)
t.left(90)
t.pendown()


def sTriangle(side, minSide):
  if (side<minSide):
    #print("side of ret: ", side)
    return
  else:
    for i in range(3):
      #print("side: ", side)
      t.forward(side)
      t.right(120)
      sTriangle(side*0.5, minSide)
  
sTriangle(200,20)
```

## Task 1

```python
#import ColabTurtle.Turtle as t
t.initializeTurtle(13, (1200, 1000))

margins = ((t.window_width() * 0.05, t.window_width() * 0.95), (t.window_height() * 0.05, t.window_height() * 0.95))

t.bgcolor('white')
t.pencolor('black')
t.width(1)
t.shape('circle')





t.penup()
t.goto(margins[0][1], margins[1][1])
#t.back(40)
t.left(90)
t.pendown()



def sTriangle(side, maxIter, iter = 0):
  if (iter>=maxIter):
    return
  else:
    for i in range(3):
      t.forward(side)
      t.right(120)
      sTriangle(side*(1/3), maxIter, iter = iter+1)
  
sTriangle((margins[1][1] - margins[1][0])* (2/1.73), 5)
```

## Task 2

```python
#import ColabTurtle.Turtle as t
t.initializeTurtle(13, (500, 500))

margins = ((t.window_width() * 0.05, t.window_width() * 0.95), (t.window_height() * 0.05, t.window_height() * 0.95))

t.bgcolor('white')
t.pencolor('black')
t.width(1)
t.shape('circle')




t.penup()
t.home()
#t.back(40)
t.right(90)
t.pendown()



def square(side, maxIter, iter = 0):
  if (iter>=maxIter):
    #t.right(90)
    return
  else:
    for i in range(2):
      t.forward(side)
      
      square(side*(1/2), maxIter, iter = iter+1)
      
      if iter%2 == 0:
        t.left(90)
      else:
        t.right(90)
      
      t.forward(side)

      if iter%2 == 0:
        t.left(90)
      else:
        t.right(90)
      
      
  
square((margins[1][1] / 6), 5)
```

```python
#import ColabTurtle.Turtle as t
t.initializeTurtle(13, (500, 500))

margins = ((t.window_width() * 0.05, t.window_width() * 0.95), (t.window_height() * 0.05, t.window_height() * 0.95))

t.bgcolor('white')
t.pencolor('black')
t.width(1)
t.shape('circle')




t.penup()
t.goto(t.window_width()/2, margins[1][1])
#t.back(40)

t.pendown()



def lines(len, maxIter, iter = 0):
  if (iter>=maxIter):

    return
  else:

    t.forward(len)

    t.left(22.5)
    lines(len*(1/2), maxIter, iter = iter+1)
    
    for i in range(2):
      t.left(45)
      lines(len*(1/2), maxIter, iter = iter+1)

    t.right(135)
    lines(len*(1/2), maxIter, iter = iter+1)

    for i in range(2):
      t.right(45)
      lines(len*(1/2), maxIter, iter = iter+1)

    t.left(90+22.5)
    t.back(len)
     
      
      
      
  
lines((margins[1][1] / 4), 4)
```
